# Training_dataset_preparation

```python
train_imgs_dir = imges_dir+"/"+"train/"
train_lbl_dir = label_dir+"/"+"train/"
```

```python
os.makedirs(train_imgs_dir,exist_ok=True)
os.makedirs(train_lbl_dir,exist_ok=True)
```

```python
def resize_img(im , target_width = 640):
    h,w,_  = im.shape
    target_height = int(h / w * target_width)
    im = cv.resize(im , (target_width , target_height), interpolation = cv.INTER_AREA)
    return im
```

```python
# Crowd human Dataset Preparation
```

```python
with open("/content/annotation_train.odgt") as fr:
  lines = fr.readlines()
  for ind,line in enumerate(lines,start=1):
    obj_dict = json.loads(line)
    # print(obj_dict)
    img_name = obj_dict['ID']
    img = cv.imread("/content/Images/{}.jpg".format(img_name))
    img_h,img_w,_ = img.shape
    with open(train_lbl_dir+"crowd_{}.txt".format(ind),mode='w') as fw:
      for gtbox in obj_dict['gtboxes']:
        if (gtbox['tag'] == "person"):
          x1,y1,w,h = gtbox['vbox']
          cx = (x1+w/2)/img_w
          cy = (y1+h/2)/img_h
          w = w / img_w
          h = h / img_h
          yolo_line = f'{0} {cx} {cy} {w} {h}\n'
          fw.write(yolo_line)
    cv.imwrite(train_imgs_dir+"crowd_{}.jpg".format(ind), resize_img(img))

```

```python
# Validation Dataset Preparation
imgs_dir = imges_dir+"/"+"val/"
lbl_dir = label_dir+"/"+"val/"
os.makedirs(imgs_dir,exist_ok=True)
os.makedirs(lbl_dir,exist_ok=True)
```

```python
with open("/content/annotation_val.odgt") as fr:
  lines = fr.readlines()
  for ind,line in enumerate(lines,start=1):
    obj_dict = json.loads(line)
    # print(obj_dict)
    img_name = obj_dict['ID']
    img = cv.imread("/content/Images/{}.jpg".format(img_name))
    img_h,img_w,_ = img.shape
    with open(lbl_dir+"crowd_val_{}.txt".format(ind),mode='w') as fw:
      for gtbox in obj_dict['gtboxes']:
        if (gtbox['tag'] == "person"):
          x1,y1,w,h = gtbox['vbox']
          cx = (x1+w/2)/img_w
          cy = (y1+h/2)/img_h
          w = w / img_w
          h = h / img_h
          yolo_line = f'{0} {cx} {cy} {w} {h}\n'
          fw.write(yolo_line)

    cv.imwrite(imgs_dir+"crowd_val_{}.jpg".format(ind), resize_img(img))
```

```python
with open("/content/train.txt") as f:
    file_names = f.readlines()
    for name in file_names:
        name = name.strip()
        img = cv.imread("/content/Images/"+name+".jpg")
        img_h,img_w,_ = img.shape

        with open("/content/Annotations/"+name+".jpg.txt") as fr:
            labels = fr.readlines()
            flag = False
            with open(train_lbl_dir+name+".txt",mode='w') as fw:
                for rect in labels[1:]:
                    cls,x1,y1,x2,y2 = rect.strip().split(" ")
                    cls,x1,y1,x2,y2 = int(cls),int(x1),int(y1),int(x2),int(y2)
                    if(cls == 1 or cls == 2 or cls == 3):
                        w = abs(x2-x1)
                        h = abs(y1-y2)
                        cx = (x1+w/2)/img_w
                        cy = (y1+h/2)/img_h
                        w = w / img_w
                        h = h / img_h
                        yolo_line = f'{0} {cx} {cy} {w} {h}\n'
                        fw.write(yolo_line)
        cv.imwrite(train_imgs_dir+name+".jpg", resize_img(img))
```
